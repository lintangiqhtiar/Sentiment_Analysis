```python
#menganalisis top 5 direktor, untuk NaN diisi director not specified
df['director']=df['director'].fillna('Director Not Specified')
df.head()
```

```python
#mengumpulkan list director kedalam data fram yang lain
#membuat data frame baru

director_list = pd.DataFrame()
print(director_list)
```

```python
director_list = df['director'].str.split(',', expand=True).stack()
print(director_list)
```

```python
director_list=director_list.to_frame()
print(director_list)
```

```python
director_list.columns=['Director']
print(director_list)
```

```python
director = director_list.groupby(['Director']).size().reset_index(name='Total Count')
print(director)
```

remove director not identified

```python
director= director[director.Director != 'Director Not Specified']
print(director)
```

diurutkan total count

```python
director = director.sort_values(by=['Total Count'], ascending = False)
print(director)
```

```python
top10Directors = director.head(10)
print(top10Directors)
```

Visualisasi

```python
top10Directors=top10Directors.sort_values(by='Total Count')
barChart = px.bar(top10Directors, x='Total Count', y='Director', title='Top 10 Director on Netflix')
barChart.show()
```

analisis yang sama namun berbeda kolom

```python
df['cast']=df['cast'].fillna('Actor Not Identified')
Actor_list=pd.DataFrame()
Actor_list=df['cast'].str.split(',',expand=True).stack()
Actor_list=Actor_list.to_frame()
Actor_list.columns=['Actor']
actors=Actor_list.groupby(['Actor']).size().reset_index(name ='Total Count')
actors=actors[actors.Actor != 'Actor Not Identified']
actors=actors.sort_values(by=['Total Count'], ascending = False)
top5Actors=actors.sort_values(by=['Total Count'])
top5Actors=actors.head(5)
BarChart2= px.bar(top5Actors, x='Total Count', y='Actor', title='TOP 5 ACTOR ON NETFLIX')
BarChart2.show()
```

analisis conten tiap tahun mulai dr tahun 2000

```python
df1 = df[['type', 'release_year']]
df1 = df.rename(columns={"release_year" : "Release Year", "type" : "Type"})
df2 = df1.groupby(['Release Year', 'Type']).size().reset_index(name = 'Total Count')
print(df2)
```

```python
df2 = df2[df2['Release Year']>=2000]
graph = px.line(df2, x='Release Year', y='Total Count', color='Type', title='Trend of Content Produced on Netflix Every Year')
graph.show()
```

membuat analisis sentimen tiap tahun

```python
df3 = df[['release_year', 'description']]
df3 = df3.rename(columns ={"release_year" : "Release Year", "description" : "Description"})

for index, row in df3.iterrows():
  d=row['Description']
  testimonial= TextBlob(d)
  p = testimonial.sentiment.polarity
  if p==0:
    sent = 'Neutral'
  elif p>0:
    sent = 'Positive'
  else:
    sent = 'Negative'
  df3.loc[[index, 2], 'Sentiment']=sent

df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name = 'Total Count')
df3
```

```python
df3 = df3[df3['Release Year']>2005]
BarGraph = px.bar(df3, x='Release Year', y='Total Count', color='Sentiment', title="Annual Sentiment Analysis of content production on Netflix")
BarGraph.show()
```
